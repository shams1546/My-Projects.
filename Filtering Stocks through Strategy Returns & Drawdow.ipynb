**IMPORTS**

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Download Data and Make it  a Data Frame.**

```python
tickers=["MSFT", "AAPL", "AMZN", "GOOG","NVDA", "UNH"]
data = yf.download(tickers, start="2020-01-01", end="2022-01-01")
data=pd.DataFrame(data)
data
```

**Visualizing their Performance through Graph for a Rough Idea**

*Tabulating the Close Values of the Stocks.*

```python
df=data.Close
df
```

*Normalising the close values of the stocks for the given time period, it becomes easier to estimate the stock which has given better returns.*

```python
close=data.loc[:,"Close"].copy()
normclose=df.div(df.iloc[0].mul(100))
normclose.plot(figsize=(12,8))
plt.legend(loc="upper left")
plt.show()
```

*Look for stocks with a consistent upward trend in their closing prices over time. Avoid stocks with erratic fluctuations or a downward trend.
Identify stocks with steady growth rather than sudden spikes.*

**SMA Predictions**

```python
#Defining Window Sizes
sma_s=50
sma_l=100
```

```python
# Create a copy to avoid modifying the original DataFrame
new_df=df.copy()
```

```python
#Calculate SMA for short and long windows
new_df["SMA_S"]=new_df.iloc[:,0].rolling(window=sma_s).mean()
new_df["SMA_L"]=new_df.iloc[:,0].rolling(window=sma_l).mean()
new_df
```

```python
#Creating the Date Column.
new_df['Date'] = new_df.index
new_df
```

```python
# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
# Plot each column in a separate subplot
for i, column in enumerate(['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'UNH']):
    row = i // 3
    col = i % 3
    axes[row, col].plot(new_df['Date'], new_df[column])
    axes[row, col].plot(new_df['Date'], new_df['SMA_S'])
    axes[row, col].plot(new_df['Date'], new_df['SMA_L'])
    axes[row, col].set_title(column)

# Plot SMA_S and SMA_L in separate subplots
axes[2, 0].plot(new_df['Date'], new_df['SMA_S'])
axes[2, 0].set_title('SMA_S')
axes[2, 1].plot(new_df['Date'], new_df['SMA_L'])
axes[2, 1].set_title('SMA_L')

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

*Trend is Bullish if (SMA_S)>(SMA_L), and Vice-Versa.*

```python
new_df.dropna(inplace=True)
new_df
```

**Creating Long Bias Strategy**

*Long Positions for Bullish trend otherwise Hold Positions*

```python
new_df["Position"]=np.where(new_df["SMA_S"]>new_df["SMA_L"],1,0)
new_df
```

```python
symbol= ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'UNH']
for i in range(len(symbol)):
  new_df["Returns of"+symbol[i]]=np.log(new_df[symbol[i]]/new_df[symbol[i]].shift(1))
new_df
```

```python
symbol= ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'UNH']
for i in range(len(symbol)):
  new_df["Strategy Returns of"+symbol[i]]=new_df["Returns of"+symbol[i]]*new_df["Position"].shift(1)
new_df.dropna(inplace=True)
new_df
```

```python
symbol= ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'UNH']
Strategy_Returns=[0,0,0,0,0,0]
for i in range(len(symbol)):
  Strategy_Returns[i]=(np.exp(new_df["Strategy Returns of"+symbol[i]].sum()))*100
  print("Strategy Returns of", {symbol[i]}, "=", {Strategy_Returns[i]})
```

**Compare Drawdowns**

*A low drawdown is generally considered good for an investment. A low drawdown signifies that the investment's value doesn't fluctuate significantly downwards from its highs.*

```python
#Calculating Cumulative Returns of each stock.
symbol= ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'UNH']
for i in range(len(symbol)):
  new_df["Cumulative Returns of"+symbol[i]]=np.exp(new_df["Strategy Returns of"+symbol[i]].cumsum())
  print(new_df["Cumulative Returns of"+symbol[i]].to_frame)
```

```python
#Calculating Cumulative Maximum of each stock.
symbol= ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'UNH']
for i in range(len(symbol)):
  new_df["Cumulative Maximum of"+symbol[i]]=new_df["Cumulative Returns of"+symbol[i]].cummax()
  print(new_df["Cumulative Maximum of"+symbol[i]].to_frame)
```

*Assuming that we have a capital of 1000$, we would calculate the maximum Drawdown encountered by each stock at the end of 2 years considering that the whole capital was exhausted in that stock.*

```python
#Calculating Maximum Drawdowns for the whole timespan
symbol= ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'UNH']
drawdown_data=[0,0,0,0,0,0]
for i in range(len(symbol)):
  new_df["Factor of"+symbol[i]]=new_df["Cumulative Returns of"+symbol[i]]+0.112723
  peak=0.112723
  max_drawdown=0.0
  for values in new_df["Factor of"+symbol[i]]:
    if values>peak:
      peak=values
    drawdown=(peak-values)/peak
    if drawdown>max_drawdown:
      max_drawdown=drawdown
  drawdown_data[i]=max_drawdown*100
  print("Maximum Drawdown of", {symbol[i]}, "=", {drawdown_data[i]})
```

**Plotting the 2 Results to be comsidered while filtering stocks**

```python
import matplotlib.pyplot as plt # Import the necessary library
plt.bar(symbol, drawdown_data) # Use plt.bar to create the plot, providing x and y data
plt.xlabel("Symbol") # Label the x-axis
plt.ylabel("Maximum Drawdown (%)") # Label the y-axis
plt.title("Maximum Drawdowns by Symbol") # Add a title
plt.show() # Display the plot
```
